# Bar Plot

```python
enrollment=[104,112,256,178,126,180]
courses = ['C','Java','Python','R','PHP','JavaScript']
```

```python
plt.xlabel('Courses')
plt.ylabel('Enrolled Student')
plt.bar(courses,enrollment,color='brown')
plt.grid()
```

```python
plt.ylabel('Courses')
plt.xlabel('Enrolled Student')
plt.barh(courses,enrollment,color='red')
plt.grid()
```

## Pie Chart

```python
plt.figure(figsize=(10,10))
plt.pie(enrollment,labels=courses,autopct='%2.2f', shadow=True)
plt.plot()
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_Data.csv')
```

## Seperate input and output data

```python
x=df['YearsExperience'].values
y=df['Salary']
```

```python
x
```

```python
y
```

```python
x=x.reshape(30,1)
```

## Import linear regression class

```python
from sklearn.linear_model import LinearRegression
```

## create object of linear regression

```python
regressor =LinearRegression()
```

### train the algorithm

```python

regressor.fit(x,y)
```

## Predict on unseen Data
